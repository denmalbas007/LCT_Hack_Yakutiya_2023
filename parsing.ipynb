```python
from tika import parser # pip install tika
def extract_text_from_pdf(path):
    raw = parser.from_file(path)
    return raw['content']
```

```python
import requests
def download_pdf(url, save_path):
    
    r = requests.get(url, stream=True)
    
    with open(save_path, 'wb') as f:
        f.write(r.content)
```

```python
import os
def pdfs2csv(df):
    texts = []
    urls = []
    
    for i in range(10):
        url = df['url'][i]
        #print(url)
        try:
            download_pdf(url, str(i)+'.pdf')
            text = extract_text_from_pdf(str(i) + '.pdf')
            text_list = text.split()
            text = ' '.join(text_list).strip()
            left = text.index("ПОСТАНОВЛЕНИЕ")
            text = text[left + 14:]
            print(text)
    
            texts.append(text)
            urls.append(url)
        
        # df = pd.concat([df, pd.DataFrame([text], columns=['text'])], ignore_index=True)
            os.remove(str(i)+'.pdf')
        except Exception:
        # df = pd.concat([df, pd.DataFrame(['error'], columns=['text'])], ignore_index=True)
            print('error')
            texts.append('error')
            urls.append(url)
        
    df = pd.DataFrame({"text": texts, "url": urls})
    
    return df
```
